```python
ruta_archivo = 'Data/btc_limp.csv' 

btc = pd.read_csv(ruta_archivo)
btc.set_index('Date', inplace=True)
```

```python

# Asegúrate de que el índice de tu DataFrame sea de tipo datetime
btc.index = pd.to_datetime(btc.index)

# Utiliza la función resample con la frecuencia 'MS' para obtener el primer día de cada mes
btc_mes = btc.resample('MS').first()

# Elimina los datos anteriores a '2014-01-01'
btc_mes = btc_mes.loc['2014-01-01':]
```

```python
btc_mes = btc_mes ['Open']
btc_mes = pd.DataFrame (btc_mes)
```

```python
btc_sin_up = btc_mes.copy ()
btc_sin_up = pd.DataFrame (btc_sin_up)
btc_sin_up.index = pd.to_datetime(btc_sin_up.index)

btc_down = btc_mes.copy ()
btc_down = pd.DataFrame (btc_down)
btc_down.index = pd.to_datetime(btc_down.index)

btc_sin_up_down = btc_mes.copy ()
btc_sin_up_down = pd.DataFrame (btc_sin_up_down)
btc_sin_up_down.index = pd.to_datetime(btc_sin_up_down.index)

btc_sin_up
```

```python
btc_mes ['retornos']= btc_mes ['Open'].pct_change ().round (2)

btc_mes ['ganancias'] = (1 + btc_mes['retornos']).cumprod()

btc_mes
```

```python
mejores = btc_mes.loc[btc_mes['retornos'].nlargest(25).index]
peores = btc_mes.loc[btc_mes['retornos'].nsmallest(25).index]

peor = btc_mes.sort_values(by='retornos', ascending=True)[:20]

peores_retornos = btc_mes['retornos'].nsmallest(25, keep='all')



peores_retornos
```

```python
btc_sin_up = btc_sin_up.drop(mejores.index)
btc_sin_up ['retornos']= btc_sin_up ['Open'].pct_change ().round (2)
btc_sin_up ['ganancias'] = (1 + btc_sin_up['retornos']).cumprod().round (2)
btc_sin_up
```

```python
btc_down = btc_down.drop(peores.index)
btc_down ['retornos']= btc_down ['Open'].pct_change ().round (2)
btc_down ['ganancias'] = (1 + btc_down['retornos']).cumprod().round (2)
btc_down
```

```python
btc_sin_up_down = btc_sin_up_down.drop(peores.index).dropna ()
btc_sin_up_down = btc_sin_up_down.drop (mejores.index).dropna ()
btc_sin_up_down ['retornos']= btc_sin_up_down ['Open'].pct_change ().round (2)
btc_sin_up_down ['ganancias'] = (1 + btc_sin_up_down['retornos']).cumprod().round (2)
btc_sin_up_down.dropna ()
```

```python
import dash
from dash import dcc, html
import plotly.graph_objs as go

# Crear la aplicación Dash
app = dash.Dash(__name__)

# Definir los datos
data = [
    {'x': btc_sin_up_down.index, 'y': btc_sin_up_down['ganancias'], 'name': 'btc_sin_up_down', 'line': {'color': 'blue', 'dash': 'solid'}},
    {'x': btc_down.index, 'y': btc_down['ganancias'], 'name': 'btc_down', 'line': {'color': 'red', 'dash': 'dash'}},
    {'x': btc_sin_up.index, 'y': btc_sin_up['ganancias'], 'name': 'btc_sin_up', 'line': {'color': 'green', 'dash': 'dot'}},
    {'x': btc_mes.index, 'y': btc_mes['ganancias'], 'name': 'btc_mes', 'line': {'color': 'orange', 'dash': 'dashdot'}}
]

# Crear el layout de la aplicación Dash
app.layout = html.Div([
    dcc.Graph(
        id='graph',
        figure={
            'data': data,
            'layout': {
                'title': 'Ganancias de Bitcoin',
                'xaxis': {'title': 'Fecha'},
                'yaxis': {'title': 'Ganancias'},
                'legend': {'orientation': 'h'}
            }
        }
    )
])

# Ejecutar la aplicación
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
# Crear un DataFrame vacío para almacenar los peores retornos de cada mes
peores_meses = pd.DataFrame(columns=btc_mes.columns)

# Iterar sobre cada mes
for fecha_inicio, datos_mes in btc_mes.groupby(pd.Grouper(freq='M')):
    # Encontrar los 20 peores retornos para este mes
    peores_mes_actual = datos_mes.nsmallest(20, 'retornos', keep='all')
    # Agregar los peores retornos de este mes al DataFrame de peores_meses
    peores_meses = pd.concat([peores_meses, peores_mes_actual])

# Ver los peores retornos para cada mes
print(peores_meses)
```

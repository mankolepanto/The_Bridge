```python
ruta_archivo = 'data/Bitcoin_Historical.csv' 

bitcoin = pd.read_csv(ruta_archivo)

bitcoin.head ()
```

```python

def convertir_a_fecha(df, columna):
    
    df[columna] = pd.to_datetime(df[columna])

    return df
convertir_a_fecha (bitcoin, 'Date')
bitcoin['Date'].apply(type)
```

```python
def convertir_columna_a_entero(df, columna):

    # Asegúrate de que la columna esté representada como cadena de caracteres
    df[columna] = df[columna].astype(str)
    
    # Elimina los puntos decimales de la columna
    df[columna] = df[columna].str.replace('.', '')
    df[columna] = df[columna].str.replace(',', '')

    # Convierte la columna a tipo entero
    df[columna] = df[columna].astype(int)

    return df
convertir_columna_a_entero (bitcoin, 'Open')
```

```python
bitcoin = bitcoin.iloc[:4000]

bitcoin.info ()
```

```python
# Log returns
btc_open = bitcoin ['Open']

btc_log_returns  = np.log (btc_open /btc_open.shift (1)).dropna ()

btc_log_returns
```

```python
# Calculando la desviacion tipica diaria

daily_std = btc_log_returns.std()

daily_std
```

```python

daily_std = np.nanstd(btc_log_returns)
print(daily_std*100)
```

```python
#Volatilidad anualizada

anualizada = daily_std * np.sqrt(365)

anualizada*100
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Crea una figura con el tamaño deseado
fig = make_subplots(rows=2, cols=2)

# Añade tu traza al subplot deseado
trace0 = go.Histogram(x=btc_log_returns, name='BTC')
fig.append_trace(trace0, 1, 1)

# Ajusta el tamaño de la figura utilizando el argumento layout
fig.update_layout( autosize = False, width=1000, height=1000)

# Muestra la figura
fig.show()
```

```python
# histograma que retorne la volatilidad anualizada

fig = make_subplots (rows = 2, cols=2)

trace0 = go.Histogram ( x = btc_log_returns, name = 'BTC')
fig.append_trace (trace0,1,1)
fig.show ()
```

```python
#trailing volatility over timy
trading_days = 60
volatility = btc_log_returns.rolling (window=trading_days).std ()*np.sqrt (trading_days)

volatility
```

```python
volatility.plot ().update_layout( autosize = False, width=600, height=300)
```

```python
#Sharpe Ratio

Rf = 0.01/365

sharpe_ratio = (btc_log_returns.rolling(window=trading_days).mean ()- Rf)*trading_days/volatility

sharpe_ratio.plot ().update_layout( autosize = False, width=600, height=300)
```

```python
sortino_vol = btc_log_returns[btc_log_returns <0].rolling (window=trading_days, center = True, min_periods = 10).std ()*np.sqrt (trading_days)

sortina_ratio = (btc_log_returns.rolling(window=trading_days).mean ()- Rf)*trading_days/sortino_vol
```

```python
sortino_vol.plot ().update_layout( autosize = False, width=600, height=300)
```
